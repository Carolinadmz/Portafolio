```python
#Cargar las librerías
import numpy as np #matrices
from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential #construir capa por capa
```

```python
#Cargar la base de datos OR2
training_data = np.array([[0,0],[0,1],[1,0],[1,1]],"float32")
```

```python
training_data.shape
```

```python
# Cargar el conjunto de etiquetas deseadas (salidas)
target_data = np.array([[0],[1],[1],[0]],"float32")
```

```python
target_data
```

```python
# Crear una red para resolver el problema de la compuerta OR2
model = Sequential()
model.add(Dense(40, input_dim=2, activation='relu')) #input_dim, es el número de entradas y 16 salidas ocultas, el valor que quiera, imput_dim solo va en la primera capa
model.add(Dense(1, activation='tanh'))
```

```python
# Compilar el modelo
model.compile(loss='mean_squared_error',
              optimizer='sgd', #sgd  descenso de gradiente estocástico
              metrics=['binary_accuracy'])
```

```python
# Entrenar el modelo
model.fit(training_data, target_data, epochs=200) #epochs son las iteraciones
```

```python
# Mostrar las predicciones
print(model.predict(training_data).round())
```
