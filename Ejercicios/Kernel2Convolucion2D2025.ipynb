```python
from skimage import io, color, exposure # simbolo de convolución *, se le aplica un kernel que es un filtro, matrices cuadradas(3x3), (4x4)...
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=io.imread('/content/drive/MyDrive/Colab Notebooks/2025-IA/perfil.jpg')
```

```python
img.shape # el 3 significa que esta en formato RGB, 3 canales: rojo, verde y azul
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img=color.rgb2gray(img)# convertir la imagen de color a grises
```

```python
plt.imshow(img, cmap=plt.cm.gray) # como ya no esta en color, se debe agregar el mapa de color
plt.axis('off')
plt.show()
```

```python
img.shape # ahora solo tiene una matriz
```

```python
def convolved2d(image, kernel):
  kernel = np.flipud(np.fliplr(kernel))
  output = np.zeros_like(image)
  image_padded = np.zeros((image.shape[0] + 2, image.shape[1]+2))
  image_padded[1:-1, 1:-1] = image
  for x in range(image.shape[1]):
    for y in range(image.shape[0]):
      output[y,x]=(kernel*image_padded[y:y+3,x:x+3]).sum()
  return output
```

```python
kernel = np.array( [ [-2,-1,0], [-1,1,1], [0,1,2] ] )
image_filt = convolved2d(img, kernel)
```

```python
plt.imshow(image_filt, cmap=plt.cm.gray) # se usa un filtro para resaltar algo
plt.axis('off')
plt.show()
```

```python
image_filt_equalized = exposure.equalize_adapthist(image_filt/np.max(np.abs(image_filt)), clip_limit=0.03)
plt.imshow(image_filt_equalized, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
